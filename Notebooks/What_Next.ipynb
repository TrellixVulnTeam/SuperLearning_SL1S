## Core Analytic Resources: 

#### Linear Algabra: Fundemental to understanding many of the algorithms
* **Beginner**:
    1. Gilbert Strang: (Book and Lectures): Not as applied but good intro to the subject
* **Advanced**: 
    1. Numerical Linear Algebra (Trefethan): Good book for numerical solutions in Linear Algebra
    
    
#### Statistics and Probability: Key in doing data science
* **Beginner**: 
    1. Open Intro Stats: Book. Sometimes there are MOOC courses that follow this book. Good introduction to statistics
    
#### Optimization: Fundemental for ML. Can be more advanced. Several parts (Convex, Non-Convext, Linear and Discrete, Combinatorial, etc)
* **Beginner**: 
    1. Linear and Discrete Optimization (Coursera: I don't know the level but looks to be **Beginner**)
* **Advanced**: 
    1. Convex Optimization: Stephen Boyd (Edx Course and Book) 
    
### Machine Learning: Fundemental for doing data science. They build off of several other areas
* **Beginner**: 
    1. Intro to Statistical Learning: Book and videos (Tibshirani and Hastie): Good intro book
    2. Coursera Machine Learning Andrew Ng
    3. Udacity Machine Learning Charles Isbell

    
* **Difficult/Advanced**: 
    1. Elements of Statistical Learning (Book: Tibshirani and Hastie)
    2. Patter Recognition and Machine Learning: Bishop
    3. Machine Learning: A Probabilistic Perspective (Murphy)
    4. Machine Learning: Tom Mitchell
    5. All of Statistics: Larry Wasserman
    6. Pedro Domingos Course
    
### Information Theory:  (Entropy, Information, etc)
* **Difficult/Advanced**:

# What I would start with


##  To Start, Beginner:

### Fundementals:
* Python for Data Analysis (Wes Mckiney): Comprehensive of Numpy and Pandas
* Intro to Statistical Learning (Tibshirani and Hastie): Their examples are in R but it's a good introduction
* Open Intro Statistics: Good foundation for statistics
* Udacity Machine Learning:(Charles Isbell) Good machine learning foundation
* Udacity Machine Learning: (Sebastian Thrun) Good intro and done in Python. Pretty Basic
* Applied Predictive Modelling: (Max Kuhn) Good book. It's in R but is very practical for real world problems
* SciPy and PyData Conference Videos

### Programming: 
* Fluent Python: Not data science specific but teaches the language
* Learning Python: Not data science specific but teaches the language
* R: Intro to Statistical Learning: Learn by Example
* R: Applied Predictive Modelling: Learn by Example
* R: Udacity Data Exploration with R: Learn by Example

### Big Data Tech:
* Udacity Data Science Course
* Edx Spark Course
* Udacity Hadoop Course

### Play with some of the visualization stuff and get a better handle on the basics of databases 

### Practice
* Try some Kaggle competitions
* Look at some Kaggle solutions and rea the forums
* Lot's of examples on blogs on stuff
* Try to find datasets and play with something you'd like

## Then Medium to Advanced

* Machine Learning A Probabilisic Perspective (Murphy): This is a great all encompassing ML book
* All of Statistics (Wasserman) and Elements of Statistical Learning (Tibshirani & Hastie): Good books on statistical machine learning with a different perspective than Murphys
* Linear Algebra Resources: Again this is a fundemental area. Focus on learning all the matrix operations, vector spaces, norms, eigenvalues, matrix factroization (SVD)
* Pedro Domingos Course: He gives great lectures
* Mining Massive Datasets

#### If you want:
* Machine Learning (Tom Mitchell): A little dated but a good intro
* Other fundemental resources, big data tech, books, lectures, etc

## More Advanced for when you get through Murphys book 

* Probabilistic Graphical Models: Book and Course (Daphne Koeller)
* Reinforcement Learning: Udacity Course with Charles Isbell
* Neural Networks for Machine Learning: Geoff Hinton Coursera
* Neural Network Book (Yoshua Bengio)
* Information Theory Mackay
