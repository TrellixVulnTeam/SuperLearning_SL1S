```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x_data = np.linspace(0, 10, 21)
y_data = [1.5*np.random.randint(0, 10) for x in x_data]
y_data
```

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)

ax.set_title("Some random numbers")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.plot(x_data, y_data, c='g')

ax.set_xlim(-1, 11)
ax.set_ylim(-1, 15)
```

```python
# Some new data, one x axis with two sets of y points:
x_data = range(20)
y1_data = [1.5 * np.random.randint(0, 10) for x in x_data]
y2_data = [1.5 * np.random.randint(0, 10) for x in x_data]
```

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)
ax.plot(x_data, y1_data, c='olive')
ax.plot(x_data, y2_data, c = 'purple')
ax.grid(True)
```

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)
ax.plot(x_data, y2_data, '+', c='brown')# Plot using red '+' signs.
ax.hist(y1_data)
```

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)
ax.scatter(x_data, y1_data, c='purple', s=[16 * y for y in y2_data])
```

```python
# Map numbers to colors:
colors = []
for y in y2_data:
    if y < 5: colors.append('purple')
    else: colors.append('g')

fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)
ax.scatter(x_data, y1_data, s=96, c=colors, linewidths=0)
```

```python
# Some more new data
x_data = range(20)
y1_data = [np.random.randint(0, 10) for x in x_data]
y2_data = [np.random.randint(0, 10) for x in x_data]
```

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)
ax.plot(x_data, y1_data, c='k', linewidth=2.0) 
ax.plot(x_data, y2_data, c='r', linestyle='--')

ax.legend(["Black line", "Dashed red line"])
```
